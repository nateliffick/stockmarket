```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv('sandp.csv')
```

```python
df
```

```python
df.head(8)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df_sorted = df.sort_values('Date', ascending=True)

df_sorted = df_sorted.reset_index(drop=True)

df_sorted.head(3)
```

```python
mean_5 = df_sorted['Close'].rolling(window=5).mean().shift()
mean_25 = df_sorted['Close'].rolling(window=25).mean().shift()
mean_50 = df_sorted['Close'].rolling(window=50).mean().shift()

# Calculating the standard deviations:
std_5 = df_sorted['Close'].rolling(window=5).std().shift()
std_25 = df_sorted['Close'].rolling(window=25).std().shift()
std_50 = df_sorted['Close'].rolling(window=50).std().shift()
```

```python
mean_5
```

```python
mean_25
```

```python
std_50
```

```python
# sort the data frame
df_sorted['mean_5'] = mean_5
df_sorted['mean_25'] = mean_25
df_sorted['mean_50'] = mean_50
df_sorted['std_5'] = std_5
df_sorted['std_25'] = std_25
df_sorted['std_50'] = std_50
```

```python
df_sorted.head(20)
```

```python
df_drop_na= df_sorted.dropna()
```

```python
df_drop_na.head(20)
```

```python
# to help predict we need to train the model

train = df_drop_na[df_drop_na['Date'] < datetime(year=2013, month=1, day=1)]
test= df_drop_na[df_drop_na['Date'] >= datetime(year=2013, month=1, day=1)]
```

```python
train.tail(15)
```

```python
train.head(15)
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
feat = [x for x in train.columns if x.startswith("mean") or x.startswith("std")]
```

```python
linearr = LinearRegression()
```

```python
linearr.fit(train[feat], train['Close'])
```

```python
predictions = linearr.predict(test[feat])
```

```python
rmse = mean_squared_error(predictions, test['Close']) ** 0.5
print("The mean square error is:", rmse)
```

```python
print(linearr.score(train[feat], train['Close']))
```

```python
df_sorted['mean_ratio'] = df_sorted['mean_25'] / df_sorted['mean_5']
df_sorted['std_ratio'] = df_sorted['std_25'] / df_sorted['std_5']

df_final = df_sorted.dropna() # Dropping NaN rows

df_final.head(10)
```

```python

train2 = df_final[df_final['Date'] < datetime(year=2013, month=1, day=1)]
test2 = df_final[df_final['Date'] >= datetime(year=2013, month=1, day=1)]
```

```python
# Using List Comprehension to generate a list of the feature columns.
features = [x for x in train2.columns if x.startswith("mean") or x.startswith("std")]

# Instantiating lr 
lr = LinearRegression()

# Training Model.
lr.fit(train2[features], train2['Close'])

# Predicting on test set
predictions = lr.predict(test2[features])

# Calculating Error
rmse = mean_squared_error(predictions, test2['Close']) ** 0.5

print("The root mean squared error is:", rmse)
```

```python
print(lr.score(train2[features], train2['Close']))
```
